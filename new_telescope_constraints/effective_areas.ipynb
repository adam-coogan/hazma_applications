```python
import numpy as np
import matplotlib.pyplot as plt

from hazma.gamma_ray_parameters import (
    effective_area_comptel,
    effective_area_egret,
    effective_area_fermi,
    effective_area_adept,
    effective_area_amego,
    effective_area_all_sky_astrogam,
    effective_area_e_astrogam,
    effective_area_gecco,
    effective_area_grams,
    effective_area_mast,
    effective_area_pangu
)
```

```python
fig_dim = 4
colors = [
    "#1f77b4",
    "#ff7f0e",
    "#2ca02c",
    "#d62728",
    "#9467bd",
    "#8c564b",
    "#e377c2",
    "#7f7f7f",
    "#bcbd22",
    "#17becf",
]
```

```python
plt.figure(dpi=150, figsize=(7, 3.5))

e_gams = np.geomspace(0.1, 1e3, 2000)

plt.plot(
    e_gams,     
    effective_area_comptel(e_gams),  
    label='COMPTEL',
    linestyle="-",
    color="k",
)

plt.plot(
    e_gams,     
    effective_area_egret(e_gams),  
    label='EGRET',
    linestyle="--",
    color="k",
)

plt.plot(
    e_gams,     
    effective_area_fermi(e_gams),  
    label='Fermi',
    linestyle=":",
    color="k",
)

plt.plot(
    e_gams,     
    effective_area_amego(e_gams),  
    label='AMEGO',
    linestyle="-",
    color=colors[2],
)
plt.plot(
    e_gams,     
    effective_area_adept(e_gams),  
    label='AdEPT',
    linestyle="-",
    color=colors[3],
)
plt.plot(
    e_gams,     
    effective_area_all_sky_astrogam(e_gams),  
    label='AS-ASTROGAM',
    linestyle="-",
    color=colors[4],
)
plt.plot(
    e_gams,     
    effective_area_gecco(e_gams),  
    label='GECCO',
    linestyle="-",
    color=colors[5],
)
plt.plot(
    e_gams,     
    effective_area_grams(e_gams),  
    label='GRAMS',
    linestyle="-",
    color=colors[6],
)
# plt.plot(
#     e_gams,     
#     effective_area_mast(e_gams),  
#     label='MAST',
#     linestyle="-",
#     color=colors[7],
# )
plt.plot(
    e_gams,     
    effective_area_pangu(e_gams),  
    label='PANGU',
    linestyle="-",
    color=colors[8],
)

plt.xlim(e_gams[[0, -1]])
plt.ylim(1e1, 1e4)
plt.xscale("log")
plt.yscale("log")
plt.gca().xaxis.set_ticks_position("both")
plt.gca().yaxis.set_ticks_position("both")
plt.gca().tick_params(axis="both", which="both", direction="in")
plt.xlabel(r"$E_\gamma$ [MeV]")
plt.ylabel(r"$A_\mathrm{eff}$ [cm$^2$]")

box = plt.gca().get_position()
plt.gca().set_position([box.x0, box.y0, box.width * 0.8, box.height])
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5), fontsize=7)

plt.savefig("figures/effective_areas_matching.pdf", bbox_inches="tight")
```
